```python
import random
import string

# function to generate password
def generate_password(length, include_special_chars=False, include_numbers=False):
    # create a list of characters to choose from
    chars = string.ascii_letters
    if include_special_chars:
        chars += string.punctuation
    if include_numbers:
        chars += string.digits
    # generate random password
    password = ''.join(random.choice(chars) for i in range(length))
    return password
```

```python
import random
import string

def generate_random_password(length, include_digits=True, include_special_chars=True):
    characters = string.ascii_letters
    if include_digits:
        characters += string.digits
    if include_special_chars:
        characters += string.punctuation

    password = ''.join(random.choice(characters) for _ in range(length))
    return password

def main():
    length = int(input("Enter the length of the password: "))
    include_digits = input("Include digits? (yes/no): ").lower() == "yes"
    include_special_chars = input("Include special characters? (yes/no): ").lower() == "yes"

    password = generate_random_password(length, include_digits, include_special_chars)
    print("Generated password:", password)

if __name__ == "__main__":
    main()
```
